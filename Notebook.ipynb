Visualisasi data membantu mengubah data yang kompleks menjadi data yang lebih mudah dipahami dan dapat diidentifikasikan pola, tren, dan anomali dalam bentuk visual seperti grafik, bagan, atau peta agar mempercepat pengambilan keputusan.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.replace(-200, np.nan, inplace=True)
df.isna().sum()
```

```python
df = df.dropna()  
df = df.fillna(df.mean(numeric_only=True))
```

```python
df.duplicated().sum()
df = df.drop_duplicates()
```

## Distribusi variabel utama dalam dataset

```python
main_vars = ["T", "CO(GT)", "NO2(GT)"]

plt.figure(figsize=(12,4))
for i, col in enumerate(main_vars, 1):
    plt.subplot(1,3,i)
    plt.hist(df[col].dropna(), bins=30)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

Hasil histogram memberikan gambaran mengenai variabel utama pada data. pada tabel ditunjukkan bahwa variabel suhu udara (T) memiliki distribusi yang lebih merata. sementara itu, distribusi rata-rat suhu udara lebih mendekati rata rata daripa suhu udara. Selain itu, variabel polutan seperti CO dan NO2 menunjukkan distribusi yang tidak simetris(skewed). Hal ini menunjukkan ada cukup variasi serta rentang luas pada suhu udara. Visualisasi ini memberikan petunjuk yang lebih membantu mengenai variasi rentang data dan karakteristik distribusi data pada polutan CO dan NO2.

## Faktor yang paling berpengaruh terhadap target

```python
target = "T"

corr = df.corr(numeric_only=True)[target].sort_values(ascending=False)

plt.figure(figsize=(6,4))
corr.drop(target).plot(kind="bar")
plt.title(f"Correlation with {target}")
plt.ylabel("Correlation Value")
plt.show()
```

Hasil analisis korelasi terlihat bahwa variabel yang paling berpengaruh terhadap suhu udara(T) yaitu RH (Relative Humidity). Hal ini menunjukkan bahwa peningkatan suhu dapat menyebabkan penurunan kelembapan relatif.

## Model Regresi Sederhana

Model yang digunakan adalah regresi linear sederhana.
Model ini bertujuan uuntuk mendukung proses analisis data dan memahami hubungan antar variabel.

Pada model ini terdapat;
Variabel independen: CO(GT) dan NO2(GT)
Variabel dependen: Suhu (T)
Regresi linear dipilih karena mudah diinterpretasikan, dan cocok untuh analisis hubungan linear

```python
X = df[['CO(GT)', 'NO2(GT)']].values
y = df['T'].values

X_b = np.c_[np.ones((X.shape[0], 1)), X]

theta = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y

print("Intercept:", theta[0])
print("Koefisien CO:", theta[1])
print("Koefisien NO2:", theta[2])
```

Hasil model menghasilkan persamaan regresi sebagai berikut:

T=8.710.43CO(GT)+0.08NO2(GT)

Intercept (8.71)
Nilai intercept menunjukkan perkiraan suhu ketika nilai CO dan NO2 berada pada nilai nol.
Nilai ini berfungsi sebagai titik awal model dan tidak menjadi fokus utama interpretasi, mengingat kondisi tersebut jarang terjadi secara nyata.
    
Koefisien CO(GT) = -0.43
Koefisien negatif menunjukkan bahwa peningkatan konsentrasi CO cenderung diikuti oleh penurunan suhu.
Setiap kenaikan satu satuan CO diperkirakan menurunkan suhu sekitar 0.43 satuan, dengan asumsi variabel lain konstan.
    
Koefisien NO2(GT) = 0.08
Koefisien positif menunjukkan bahwa peningkatan NO2 cenderung diikuti oleh peningkatan suhu, meskipun pengaruhnya relatif kecil.
Setiap kenaikan satu satuan NO2 diperkirakan meningkatkan suhu sekitar 0.08 satuan.
